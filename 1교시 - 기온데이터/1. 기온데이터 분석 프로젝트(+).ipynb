```python
import matplotlib.pyplot as plt
import csv
result = []
year_list = []
f = open('seoul_180803.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)
year = 1907
s = 0
cnt = 0
for row in data :
    if row[-3] != '' :
        cnt += 1
        s += float(row[-3])
    if row[0].split('-')[0] != year and cnt > 360:
        result.append(s/cnt)
        year_list.append(year)
        year = row[0].split('-')[0]
        s = 0
        cnt = 0
plt.style.use('ggplot')
plt.figure(dpi=300)
plt.plot(year_list, result)
plt.show()
```

```python
# 역대 가장 더웠던 1주일은 언제일까?

import csv
f = open('seoul_180803.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)
data = list(data)

max_date = []
max_record = []

for i in range(len(data)) :
    date = []
    record = []
    for j in range(7) :
        if data[i+j][-1] == '' :
            break
        date.append(data[i+j][0])
        record.append(float(data[i+j][-1]))
    if sum(record) > sum(max_record) : 
        max_record = record
        max_date = date

print(max_date, max_record)
```
